```python
"""
Import the dataset used for the binary classification
The dataset is in csv format so we will use the pandas module to read the csv file into a pandas DataFrame
"""
import pandas as pd
import os
filename="../dataset/churn.csv"
dataset=pd.read_csv(filename, sep=";")
dataset.head()
```

```python
"""
Exploratory Data Analysis

Questions to answer:
1) Distribution of customers per country
2) Distribution of customer service calls (amone those with 0,1 on target variable)
3) Customer service calls per gender
4) Distribution of customers that have International plan subscription among Loyal, Churners (target 0,1)
"""
```

```python
"""
Now that we have imported the dataset into a python DataFrame we can explore the data.
First and foremost lets identify the X features (predictors) and the target variable
Target/Dependent variable is the column: Churn which indicates whether or not the customer left the Telco company or rather stayed.
"""
x_features=dataset.loc[:,~dataset.columns.isin(["Churn"])]
y_feature=dataset.loc[:,dataset.columns.isin(["Churn"])]
```

```python
"""
Explore the X features.
1) Note 1: Sklearn module and most python modules related to data science and machine learning expect features to be numeric representation of vectors.
Thus, having objervations of type objects (strings) it's not acceptable. Those type of predictors should be transformed to numerical vectors.
2) [State, Area Code] are categorical variables. This means that their values represent a label tag. This label tag should be transform into a numerical represantation.
The most common way to handle such types of data is to use One-Hot encoding.
3) The rest of the object variables such as [Day Mins, Eve Mins, etc] are numerical type variables but due to the comma(,) decimal separator csv reader recognised them as objects.
"""
x_features.dtypes
```

```python
"""
Approach 1: Transform selected columns to type float64 using the function .astype(float)
Approach 2: Modify the thousands= argument of the pd.csr reader function.
"""
numerical_features_to_transform=["Day Mins", "Eve Mins", "Night Mins", "Intl Mins", "Day Charge", "Eve Charge", "Night Charge", "Intl Charge"]
x_features[numerical_features_to_transform] = x_features.loc[:,numerical_features_to_transform].apply(lambda x: x.str.replace(',', '').astype(float), axis=1)
```

```python
"""
The selected predictors were successfully transformed to floats.
"""
x_features.dtypes
```

```python
"""
Before proceeding to transforming the features: State, Area Code to numerical vectors it's useful to check the number of null values in our predictors.

After results: Hopefully no NA or Null values were detected. Thus, no imputation of values will be applied.
In general for imputation methods we usually prefer to replace null values with the mode or the median value of a column.
Median is mostly preferred over mean/average because the latter is affected by outliers. So, if we aren't sure about the level of outliers in our data it's safer to first try out median.
"""
x_features.isnull().sum() #approach 1
#x_features.isna().sum() #approach 2
```

```python
"""
Having checked the number of null values in our predictors its time to convert columns: State, Area code to numeric vectors.
Step 1: Count the number of unique values per column. With this step we will take a first glampse of how many new features we will create.
Step 2: Apply one-hot encoding using the pd.get_dummies()
"""
print(f'Unique states: {x_features["State"].nunique()}') #51 states. One-hot encoding the column will generate 51 new columns with values 0,1. So in total the dataset will have 51+20=71 predictors
print(f'Unique genders: {x_features["Gender"].nunique()}')
```

```python
"""
For now we will keep all the 51 states. We want the baseline classifier to be trained on all of our predictors.
But its vital later on our analysis to identify possible states that should be removed based on their contribution in predicting the target variable.
"""
dummy_features=["State", "Gender"]
x_features=pd.get_dummies(x_features, columns=dummy_features)
```

```python
"""
Moving on with the exploratory data analysis path, we should identify the scale of the numerical predictors.
In cases of high scale difference we should consider standardizing or normalizing the values of the predictors.

After results: Based on the mean value per column, many are the predictors in thousand, hundred or decimal scale.
All the predictors used to feed the baseline model should be on the same scale.
"""
x_features.describe()
```

```python
x_features.columns
```

```python
"""
Feature engineering if necessary.
The dataset give us a possibility of creating new features by exploiting the columns of Minutes, Calls and Charges.
Those three aforementioned columns are separated into Day, Eve, Night. So we can concatenate them into a union column to include the daily Minutes, Calls and Charges per customer.
"""
x_features["UnionMinutes"] = x_features["Day Mins"] + x_features["Eve Mins"] + x_features["Night Mins"]
x_features["UnionCalls"] = x_features["Day Calls"] + x_features["Eve Calls"] + x_features["Night Calls"]
x_features["UnionCharges"] = x_features["Day Charge"] + x_features["Eve Charge"] + x_features["Night Charge"]
```

```python
"""
Scale X features.
Both StandardScaler and MinMaxScaler are very sensitive to the presence of outliers.
And since we did not include any outlier detection yet we will use RobustScaler.
"""
from sklearn.preprocessing import RobustScaler, MinMaxScaler
rob_scaler = RobustScaler()
norm_scaler = MinMaxScaler()

rob_scaled_x_features = rob_scaler.fit_transform(x_features.values)
rob_scaled_x_features = pd.DataFrame(rob_scaled_x_features, columns=x_features.columns.tolist())
rob_scaled_x_features.head()

norm_scaled_x_features = norm_scaler.fit_transform(x_features.values)
norm_scaled_x_features = pd.DataFrame(norm_scaled_x_features, columns=x_features.columns.tolist())
norm_scaled_x_features.head()
```

```python
"""
An alternative and more informative approach to confusion matrix using code from this repo: 
https://github.com/DTrimarchi10/confusion_matrix/blob/master/cf_matrix.py
"""
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def make_confusion_matrix(cf,
                          group_names=None,
                          categories='auto',
                          count=True,
                          percent=True,
                          cbar=True,
                          xyticks=True,
                          xyplotlabels=True,
                          sum_stats=True,
                          figsize=None,
                          cmap='Blues',
                          title=None):
    """
    This function will make a pretty plot of an sklearn Confusion Matrix cm using a Seaborn heatmap visualization.
    Arguments
    ---------
    cf:            confusion matrix to be passed in
    group_names:   List of strings that represent the labels row by row to be shown in each square.
    categories:    List of strings containing the categories to be displayed on the x,y axis. Default is 'auto'
    count:         If True, show the raw number in the confusion matrix. Default is True.
    normalize:     If True, show the proportions for each category. Default is True.
    cbar:          If True, show the color bar. The cbar values are based off the values in the confusion matrix.
                   Default is True.
    xyticks:       If True, show x and y ticks. Default is True.
    xyplotlabels:  If True, show 'True Label' and 'Predicted Label' on the figure. Default is True.
    sum_stats:     If True, display summary statistics below the figure. Default is True.
    figsize:       Tuple representing the figure size. Default will be the matplotlib rcParams value.
    cmap:          Colormap of the values displayed from matplotlib.pyplot.cm. Default is 'Blues'
                   See http://matplotlib.org/examples/color/colormaps_reference.html
                   
    title:         Title for the heatmap. Default is None.
    """

    # CODE TO GENERATE TEXT INSIDE EACH SQUARE
    blanks = ['' for i in range(cf.size)]

    if group_names and len(group_names)==cf.size:
        group_labels = ["{}\n".format(value) for value in group_names]
    else:
        group_labels = blanks

    if count:
        group_counts = ["{0:0.0f}\n".format(value) for value in cf.flatten()]
    else:
        group_counts = blanks

    if percent:
        group_percentages = ["{0:.2%}".format(value) for value in cf.flatten()/np.sum(cf)]
    else:
        group_percentages = blanks

    box_labels = [f"{v1}{v2}{v3}".strip() for v1, v2, v3 in zip(group_labels,group_counts,group_percentages)]
    box_labels = np.asarray(box_labels).reshape(cf.shape[0],cf.shape[1])


    # CODE TO GENERATE SUMMARY STATISTICS & TEXT FOR SUMMARY STATS
    if sum_stats:
        #Accuracy is sum of diagonal divided by total observations
        accuracy  = np.trace(cf) / float(np.sum(cf))

        #if it is a binary confusion matrix, show some more stats
        if len(cf)==2:
            #Metrics for Binary Confusion Matrices
            precision = cf[1,1] / sum(cf[:,1])
            recall = cf[1,1] / sum(cf[1,:])
            f1_score = 2*precision*recall / (precision + recall)
            stats_text = " ({}) \n\nAccuracy={:0.3f}\nPrecision={:0.3f}\nRecall={:0.3f}\nF1 Score={:0.3f}".format(
                categories[1],accuracy,precision,recall,f1_score)
        else:
            stats_text = "\n\nAccuracy={:0.3f}".format(accuracy)
    else:
        stats_text = ""


    # SET FIGURE PARAMETERS ACCORDING TO OTHER ARGUMENTS
    if figsize==None:
        #Get default figure size if not set
        figsize = plt.rcParams.get('figure.figsize')

    if xyticks==False:
        #Do not show categories if xyticks is False
        categories=False


    # MAKE THE HEATMAP VISUALIZATION
    plt.figure(figsize=figsize)
    sns.heatmap(cf,annot=box_labels,fmt="",cmap=cmap,cbar=cbar,xticklabels=categories,yticklabels=categories)

    if xyplotlabels:
        plt.ylabel('True label')
        plt.xlabel('\nPredicted label' + stats_text)
    else:
        plt.xlabel(stats_text)
    
    if title:
        plt.title(title)
```

```python
"""
Calculate false negative ratio.
The false negative rate is the proportion of the individuals with a known positive condition for which the test
result is negative. 
This rate is sometimes called the miss rate.

After results: The false negative rate of the baseline classifier is relatively high. This means that our classifier missed over 80% of the Churn customers.
"""
def fnr(confusion_matrix_predictions, model_name):
    """
    Estimate the false negative rate of the model classifier defined by the parameter model_name
    --Arguments:
        confusion_matrix_predictions: the confusion matrix generated from model predictions vs true values. Matrix object generated from confusion_matrix() method of sklearn.
        model_name: a string tag that will represent the fnr ratio of the current model (i.e svm, logistic, random_forest, etc). A random identification string type object.
    """
    tn, fp, fn, tp = confusion_matrix_predictions.ravel()
    fnr_rate = fn/(tp+fn)
    print(f"{model_name} - true positives: {tp}")
    print(f"\n{model_name} - true negatives: {tn}")
    print(f"\n{model_name} - false positives: {fp}")
    print(f"\n{model_name} - false negatives: {fn}")
    
    return round(fnr_rate,4)

print(f"\nFalse positive rate of baseline classifier: {fnr(mat, 'logistic_regression_baseline')}")
```

```python
dictionary_of_comparison_scores = {}
dictionary_of_comparison_scores |= {'logistic-baseline-fnr': fnr(mat, 'logistic_regression_baseline'),
                                    'logistic-baseline-recall': round(recall_score(y_test.ravel(), y_pred),4)
                                   }
dictionary_of_comparison_scores
```

```python
"""
Calculate the bias and variance of the classifier. Bias-variance tradeoff will help us to identify if the dataset is over or under fitted
After result: Both bias and variance converge towards zero. This is a good sign about the overfit or underfit of the model classifier.
Even though bias-variance tradeoff is a good measure for model evaluate and generalization. Here its power is low due to
great imbalance in the dataset. The baseline classifier tends to favor the negative -0- class by achieving a high accuracy.
However, since we value the most False Negative ratio and recall. Baseline classifier has poor predictions over the positive class -1- Churn customers.
"""
def get_bias(predicted_values, true_values):
    """ Calculates model bias:param predicted_values: values predicted by the model
    :param true_values: true target values for the data
    :return: integer representing the bias of the model
    """

    return np.round(np.mean((predicted_values - true_values) ** 2), 4)

def get_variance(values):
    """ Calculates variance of an array of values:param values: array of values
    :return: integer representing the variance of the values
    """

    return np.round(np.var(values), 0)

print(f"baseline model bias: {get_bias(y_pred, y_test.ravel())}")
print(f"baseline model variance: {get_variance(y_pred)}")
```

```python
"""
So far, to evaluate the baseline model classifier we used the following methods:

1) Classification report to calculate Accuracy, Recall, Precision, F1-score (harmonic mean of Precision and Recall)
2) False Negative Ratio
3) Bias-Variance tradeoff

Binary classification cannot stop on those three evaluation metrics. Based on our case of greately imbalanced cases of positive labels we can exploit and evaluate the model classifier using the following two measures:
4) Precision-Recall Curve
5) F2-Measure
Those two extra measures come in handy to evaluate the performance of a binary classifier when the positive class is more important than the negative class

ROC AUC vs Precision Recall curve:
A ROC curve is a diagnostic plot for summarizing the behavior of a model by calculating the false positive rate and true positive rate for a set of predictions by the model under different thresholds.
Each threshold is a point on the plot and the points are connected to form a curve. 
A classifier that has no skill (e.g. predicts the majority class under all thresholds) will be represented by a diagonal line from the bottom left to the top right.

F-measure vs Fbeta-measure
The F-measure balances the precision and recall.

On some problems, we might be interested in an F-measure with more attention put on precision, such as when false positives are more important to minimize, but false negatives are still important.

On other problems, we might be interested in an F-measure with more attention put on recall, such as when false negatives are more important to minimize, but false positives are still important.
The solution is the Fbeta-measure.

Fbeta = ((1 + beta^2) * Precision * Recall) / (beta^2 * Precision + Recall)
this beta value adjust the importance on precision and recall.

Three common values for the beta parameter are as follows:

F0.5-Measure (beta=0.5): More weight on precision, less weight on recall.
F1-Measure (beta=1.0): Balance the weight on precision and recall.
F2-Measure (beta=2.0): Less weight on precision, more weight on recall.

After result: As we expected, the F2-measure is low because it favors recall and False Negatives. Since FNR is low, so we would expect f2-measure to be. Thus, we should try to fix this value.
"""

from sklearn.metrics import fbeta_score, recall_score, precision_score
    
f2_measure_baseline_clf = fbeta_score(y_true = y_test.ravel(), y_pred = y_pred, beta = 2.0)
precision_clf = precision_score(y_test.ravel(), y_pred)
recall_clf = recall_score(y_test.ravel(), y_pred)
print('Result-F2 score: precision=%.3f, recall=%.3f, fbeta=%.3f' % (precision_clf, recall_clf, f2_measure_baseline_clf))

from sklearn.metrics import plot_precision_recall_curve, auc, precision_recall_curve
import matplotlib.pyplot as plt

# Plot the Precision-Recall Curve
plt.rcParams["figure.figsize"] = (18,8)
disp = plot_precision_recall_curve(fit_model, X_test, y_test.ravel())
disp.ax_.set_title('2-class Precision-Recall curve: ')

# Calculate the Rrecision-Recall auc score
y_proba = fit_model.predict_proba(X_test)
positive_class_probs = y_proba[:, 1] #positive label are the Churners.
precision, recall, _ = precision_recall_curve(y_test.ravel(), positive_class_probs)
auc_score = auc(recall, precision)
print('Baseline PR AUC: %.3f' % auc_score)
dictionary_of_comparison_scores |= {'logistic-baseline-pr auc': round(auc_score,4),
                                    'logistic-baseline-f2-measure': round(fbeta_score(y_true = y_test.ravel(), y_pred = y_pred, beta = 2.0),4)}
dictionary_of_comparison_scores
```

n = number of features,
m = number of training examples

1. If n is large (1–10,000) and m is small (10–1000) : use logistic regression or SVM with a linear kernel.

2. If n is small (1–10 00) and m is intermediate (10–10,000) : use SVM with (Gaussian, polynomial etc) kernel.

3. If n is small (1–10 00), m is large (50,000–1,000,000+): first, manually add more features and then use logistic regression or SVM with a linear kernel.
Generally, it is usually advisable to first try to use logistic regression to see how the model does.
If it fails then you can try using SVM without a kernel (is otherwise known as SVM with a linear kernel).
Logistic regression and SVM with a linear kernel have similar performance but depending on your features, one may be more efficient than the other.

Scenario path 1: Train a more suited algorithm using hyper parameter tuning, keeping the same number of features.
-> Train Logistic Regression by using a different solver instead of the default one ('lbfgs').
-> Train Support Vectors Machines with hyper parameter tuning to identify the parameters that will identify the best model classifier.

Hyper-parameter tuning is a very useful tool to train and test different parameters per model classifier. With hyper-parameter tuning we can test different parameter combinations and select the one that best fits our data.
Cross Validation is a very useful technique to include an out-of-fold dataset that will be used to evaluate the ability of the model classifier to perfectly distinguish the churners -positive class- from the loyal -negative class- customers.

Cross validation is a sophisticated method to train and test model estimators in machine learning that will outplay the predictions of a baseline model.
Also to make thinks more compact we can add them into a pipeline object offered by sklearn library.

* Pipelines will be a concepts developed later to make our machine learning solution more compact. For we develop each step individually.

```python
def calculate_update_measures(model_name, model_trained, y_true, model_predictions, positive_class_probabilities, metrics_dictionary):
    """
    Calculate and update the dictionary of the evaluation metrics.
    This dictionary is the starting point of comparing the trained model classifiers.
    --Argumetns: 
        model_name: an identifier of the metrics calculated (i.e svm, random_forest, logistic_regression, etc). Any value of type string
        model_trained: GridSearch object that describes the best model classifier.
        y_true: the unknown sample of values, never-seen-before from the classifier.
        model_predictions: predicted values from the never seen before dataset.
        positive_class_probabailities: the probabilities of churners (positive class) from the unkown sample.
        metrics_dictionary: the previous instance of the evaluation metrics dictionary.
    --Return:
        Updated dictionary of evaluation metrics.
    """
    import json
    from sklearn.metrics import confusion_matrix, recall_score, auc, precision_recall_curve, fbeta_score
    try:
        y_true = y_true.ravel()
    except:
        y_true = y_true

    precision, recall, _ = precision_recall_curve(y_true, positive_class_probabilities)
    auc_score = auc(recall, precision)
    confusion_matrix_values = confusion_matrix(y_true, model_predictions)
    
    assert type(dictionary_of_comparison_scores) is dict, "The data structure that includes the evaluation metrics must be a dictionary."
    
    metrics_dictionary |= {
                            f'{model_name}-fnr': fnr(confusion_matrix_values, model_name),
                            f'{model_name}-recall': round(model_trained.best_score_,4),
                            f'{model_name}-pr auc': round(auc_score,4),
                            f'{model_name}-f2-measure': round(fbeta_score(y_true = y_true, y_pred = model_predictions, beta = 2.0),4)
                           }
    print(f'\n{model_name} model bias: {get_bias(model_predictions, y_true)}',
          f'\n{model_name} model variance: {get_variance(model_predictions)}')
    print(json.dumps(metrics_dictionary, indent=4))

    return metrics_dictionary
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedShuffleSplit, GridSearchCV

parameters_grid = {
                    'C':[0.001, 0.01, 0.5, 1.0, 1.5],
                    'gamma':[0.001, 0.01, 0.1, 0.5],
                    'kernel':('linear', 'rbf', 'poly')
                  }
scoring = {
            'accuracy': 'accuracy',
            'f1_scoring': 'f1',
            'precision_scoring': 'precision',
            'recall_scoring': 'recall'
          }
refit_score = list(scoring.keys())[list(scoring.values()).index('recall')]
#create stratified and shuffled cross validation rounds
ss = StratifiedShuffleSplit(n_splits=5, test_size=test_split_ratio, random_state=123)
#create the gridsearch object of the svm classifier that will be fitted on the data
svm_gs = GridSearchCV(estimator=SVC(random_state=123, probability=True),
                      param_grid=parameters_grid,
                      cv = ss,
                      verbose=1,
                      n_jobs = -1,
                      scoring=scoring,
                      refit="recall_scoring",
                      return_train_score=True)
#fit the gridsearch object
svm_gs.fit(X_train, y_train.ravel())
print(f'\nbest parameters: {svm_gs.best_params_}')
#make predictions
y_pred=svm_gs.best_estimator_.predict(X_test)
positive_class_probs = svm_gs.best_estimator_.predict_proba(X_test)[:, 1] #positive label are the Churners.

#calculate the new values of the comparison measures and enrich the dictionary.
#prerequisits: model predictions (.predict), positive class probabilities (.predict_proba).
dictionary_of_comparison_scores = calculate_update_measures(model_name="svm", 
                                                            model_trained=svm_gs, 
                                                            y_true=y_test, 
                                                            model_predictions=y_pred, 
                                                            positive_class_probabilities=positive_class_probs, 
                                                            metrics_dictionary=dictionary_of_comparison_scores)
results = svm_gs.cv_results_
results_dataframe = pd.DataFrame.from_dict(results)
best_model_index = results['params'].index(svm_gs.best_params_)

# Plot the Precision-Recall Curve
plt.rcParams["figure.figsize"] = (18,8)
disp = plot_precision_recall_curve(svm_gs, X_test, y_test.ravel())
disp.ax_.set_title('2-class Precision-Recall curve: ')

# add checkpoint logs when a model is trained successfully and in general when an important checkpoint is reached.
```

Some extra steps that we can experiment on to improve the model are feature selection techniques. Those techniques will try to identify if all the 74 features that have been used to train the three classifiers are more than the appropriate number of features that actually need to predict correctly all the positive values.
<b>Feature selection</b> is comonly used when the number of features is greately large. Here 74 features is not that many features however it's worth to try and use such techniques here. Also, feature selection can help the researcher to evaluate the importance of its feature independently and estimate their contribution to y target.

Feature selection can have the following effects:
* Improve model fit times
* Improve model metrics by including only the important features and drop those that negatively affect y target
* Reduce model complexity if the model classifier is overfitting

Feature selection that we will try out:
* Feature importance from Random Forest classifier
* PCA

```python
"""
Impurity based importance. This means that the importance of a feature is measured by its contribution in the mean decrease of impurity inside a tree.
The higher the impurity decrease the more important the feature.
In general sklearn deployes three types of impurity:
1) Gini impurity (default option in classification)
2) Entropy (second option in classification)
3) Variance (regression model)

Impurity measures the homogeneity of labels inside a tree node. The less the impurity the more homogenious the sample (only positive class observation or pure negative class observations) and purer results of a classified label in this node.

Cons of impurity-based feature importance:
1) Bias over high cardinality features (i.e features with many unique labels).
   One such feature is the column of country, each customer belongs.
2) Impurity-based importances are computed on training set statistics and therefore do not reflect the ability
   of feature to be useful to make predictions that generalize to the test set (when the model has enough capacity).

To tackle those two disadvantage we can deploy the permutation feature importance included in the inspection module of sklearn.


https://towardsdatascience.com/an-implementation-and-explanation-of-the-random-forest-in-python-77bf308a9b76
https://bambielli.com/til/2017-10-29-gini-impurity/
"""
feat_importances = pd.Series(gridsearch_classifier.best_estimator_.feature_importances_, index=x_features.columns)
feat_importances.nlargest(25).plot(kind='barh')
```

```python
fi = pd.DataFrame({'feature': list(x_features.columns),
                   'importance': gridsearch_classifier.best_estimator_.feature_importances_}).\
                    sort_values('importance', ascending = False)
fi.head(25) #top 25 features should be a user-defined value.a
```

```python
"""
Permutation based feature importance
In this example we use the test sample (never seen by random classifier to estimate the feature importance)

https://scikit-learn.org/stable/modules/generated/sklearn.inspection.permutation_importance.html#sklearn.inspection.permutation_importance
https://scikit-learn.org/stable/auto_examples/inspection/plot_permutation_importance.html#sphx-glr-auto-examples-inspection-plot-permutation-importance-py
https://scikit-learn.org/stable/auto_examples/ensemble/plot_forest_importances.html#sphx-glr-auto-examples-ensemble-plot-forest-importances-py
"""
from sklearn.inspection import permutation_importance
result = permutation_importance(gridsearch_classifier, X_test, y_test.ravel(), n_repeats=10, random_state=42, n_jobs=-1)

sorted_idx = result.importances_mean.argsort()[-25:][::-1] #top 25 most important features.

fig, ax = plt.subplots()
ax.boxplot(result.importances[sorted_idx].T,
           vert=False, 
           labels=x_features.columns[sorted_idx])
ax.set_title("Permutation Importances (test set)")
fig.tight_layout()
plt.show()
```

```python
permutated_fi = pd.DataFrame({'feature': list(x_features.columns[sorted_idx]),
                              'importance': result.importances_mean[sorted_idx].T}).\
                    sort_values('importance', ascending = False)
permutated_fi
```

But which one should we keep? UnionCharges or Day Charge? To answer this question we will consult the second dataframe. From there we can see that UnionCharges has almost the same effect on target as DayCharge column. This is expected since UnionCharges was feature engineered from Day,Eve,Night charges. Even though almost same impact on target label, we will keep UnionCharges column and we will drop DayCharge column.</div>

```python
"""
Those are the 5 features to be kept from each highly correlated pair of features. 
The opposite pair will be dropped from the top 25 most important features.
"""
important_features = set([correlation_target.index[correlation_target['target']==np.maximum(correlation_target.loc[index_pair.strip('()').split(',')[0]].values[0], correlation_target.loc[index_pair.strip('()').split(',')[1]].values[0])].values[0] if (correlation_target.loc[index_pair.strip('()').split(',')[0]].values[0]>0 and correlation_target.loc[index_pair.strip('()').split(',')[1]].values[0]>0) or (correlation_target.loc[index_pair.strip('()').split(',')[0]].values[0]*correlation_target.loc[index_pair.strip('()').split(',')[1]].values[0]<0) else correlation_target.index[correlation_target['target']==np.minimum(correlation_target.loc[index_pair.strip('()').split(',')[0]].values[0], correlation_target.loc[index_pair.strip('()').split(',')[1]].values[0])].values[0] for index_pair in correlated_features.index])
remove_correlated_features = correlation_target[~correlation_target.index.isin(list(important_features))].index.values.tolist()
remove_correlated_features
```

```python
"""
The final most important features to keep and re-train random forest classifier. Also, the other two previous
classifiers maybe be re-trained. Although, for now we will re-train only the Random Forest one.

So we will extract those features to filter out the initial dataset of x_features.
"""
fi_top = fi_top[~fi_top['feature'].str.contains('^State|^Area|^Gender')]
fi_top = fi_top[~fi_top['feature'].isin(remove_correlated_features)]
keep_columns = fi_top['feature'].values.tolist()
keep_columns
```

```python
from sklearn.ensemble import RandomForestClassifier #https://towardsdatascience.com/an-implementation-and-explanation-of-the-random-forest-in-python-77bf308a9b76

#add execution triggers of the script. Like a time scheduler.
#this could be a json file given by the user before start training
parameters_grid = {
                    'n_estimators':[25, 50, 100, 200, 250, 500, 1000],
                    'max_samples':[5, 10, 20, 40, 0.25, 0.5, 0.75], #int or float (between 0,1)
                    'oob_score':[True, False]
                  }
#this could be a json file given by the user before start training
scoring = {
            'accuracy': 'accuracy',
            'f1_scoring': 'f1',
            'precision_scoring': 'precision',
            'recall_scoring': 'recall'
          }
refit_score = list(scoring.keys())[list(scoring.values()).index('recall')]
#create stratified and shuffled cross validation rounds
ss = StratifiedShuffleSplit(n_splits=5, test_size=test_split_ratio, random_state=123)
#create the gridsearch object of the random_forest classifier that will be fitted on the data
gridsearch_classifier = GridSearchCV(estimator=RandomForestClassifier(random_state=123),
                                     param_grid=parameters_grid,
                                     cv=ss,
                                     verbose=1,
                                     n_jobs=-1,
                                     scoring=scoring,
                                     refit="recall_scoring", #in a visual interface select refit from the available options of the scoring dictionary.
                                     return_train_score=True)
#fit the gridsearch object
gridsearch_classifier.fit(X_train, y_train.ravel())
print(f'\nbest parameters: {gridsearch_classifier.best_params_}')
#make predictions
y_pred=gridsearch_classifier.best_estimator_.predict(X_test)
positive_class_probs = gridsearch_classifier.best_estimator_.predict_proba(X_test)[:, 1] #positive label are the Churners.

#calculate the new values of the comparison measures and enrich the dictionary.
#prerequisits: model predictions (.predict), positive class probabilities (.predict_proba).
dictionary_of_comparison_scores = calculate_update_measures(model_name="random_forest-top25", 
                                                            model_trained=gridsearch_classifier, 
                                                            y_true=y_test, 
                                                            model_predictions=y_pred, 
                                                            positive_class_probabilities=positive_class_probs, 
                                                            metrics_dictionary=dictionary_of_comparison_scores)
results = gridsearch_classifier.cv_results_
results_dataframe = pd.DataFrame.from_dict(results)
best_model_index = results['params'].index(gridsearch_classifier.best_params_)

# Plot the Precision-Recall Curve
plt.rcParams["figure.figsize"] = (18,8)
disp = plot_precision_recall_curve(gridsearch_classifier, X_test, y_test.ravel())
disp.ax_.set_title('2-class Precision-Recall curve: ')
```

Moving on, we will experiment on two different model selection techinques:
* Princial component analysis (PCA)

When/Why to use PCA?

a) PCA technique is particularly useful in processing data where multi-colinearity exists between the features/variables.

b) PCA can be used when the dimensions of the input features are high (e.g. a lot of variables).

c) PCA can be also used for denoising and data compression.

https://towardsdatascience.com/pca-clearly-explained-how-when-why-to-use-it-and-feature-importance-a-guide-in-python-7c274582c37e

```python
from sklearn.decomposition import PCA
pca_fs = PCA(n_components = 25, random_state=123).fit(rob_scaled_x_features)
pca_fs.explained_variance_ratio_
```

```python
variance_counter = 0.0
number_comp = 0.0
for component in range(len(pca_fs.explained_variance_ratio_)):
    while variance_counter <= 0.9:
        number_comp += 1
        variance_counter += pca_fs.explained_variance_ratio_[component]
        print(variance_counter)
number_comp
```

```python
"""
Those are the columns that have the greatest importance inside of each PCA components.
We selected the first 8 PCA components since they explain the 90% of the variance.
From those 8 features we will remove the highly correlated ones, and we will refit the Random Forest classifier.
Recall, that the highly correlated features have been already identified from the heatmap created earlier.
"""
n_pcs = int(number_comp)
most_important = [np.abs(pca_fs.components_[i]).argmax() for i in range(n_pcs)]
most_important_names = [rob_scaled_x_features.columns.tolist()[most_important[i]] for i in range(n_pcs)]
dic = {'PC{}'.format(i): most_important_names[i] for i in range(n_pcs)}
df = pd.DataFrame(dic.items(), columns=["Compontent", "MostImportantFeaturePerComp"])
df
```

```python
columns_fit_pca = df['MostImportantFeaturePerComp'][~df['MostImportantFeaturePerComp'].isin(remove_correlated_features)].values.tolist()
columns_fit_pca
```

```python
"""
Train and evaluate random forest algorithm using the top most important features (MDI > 0.1).
"""
test_split_ratio = 0.20
X_train, X_test, y_train, y_test = train_test_split(x_features[columns_fit_pca].values, y_feature.values, test_size=test_split_ratio, random_state=123, shuffle=True, stratify=y_feature)
print(f"Label ratio in y train: {[round(ele/y_train.shape[0]*100,2) for ele in np.unique(y_train, return_counts=True)[1].tolist()]}")
print(f"Label ration i y test: {[round(ele/y_test.shape[0]*100,2) for ele in np.unique(y_test, return_counts=True)[1].tolist()]}")
```
